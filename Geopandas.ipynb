```python
import geopandas as gpd

import pandas as pd
```

```python
url = 'export (1).geojson'
districts = gpd.read_file(url, crs="EPSG:4326")


districts.to_crs(epsg=3406, inplace=True)
districts.crs



    
```

```python
districts = districts[districts["id"].apply(
    lambda x: x.startswith("relation")
)]
ax= districts.plot(figsize=(10,6))

```

```python
districts.area
```

```python
districts.name
districts.loc[:,"name_lower"] = districts.name.apply(lambda x: str(x).lower())
```

```python

df = pd.read_csv("./data/Processed/ds_toado_signal_final_10_11_2023.csv")
df.loc[:,"name_lower"] = df[["ADM3","Cấp"]].apply(
    lambda x: f'{x["Cấp"]} {x["ADM3"]}'.lower(),
    axis=1
)
```

```python
df
```

```python
m = districts.merge(df, on="name_lower", how="inner")
```

```python
grouped = m.groupby("@id")["label"]
m[grouped.transform(lambda x: x.count())==1]

```

```python
m.loc[:, "signal_level"] = m["level.1"].apply(
    lambda x: -int(x[-1]) if x[-1] != 0 else -5
)
```

```python
m["level.1"].unique()
```

```python
m.loc[:, "density"] = m["Pop"]/m.area
```

```python
m.plot(cmap='Blues',linewidth=0.2,# scheme='quantiles',
    edgecolor='gray',column='signal_level',
    legend=True,figsize=(100, 60),
    #legend_kwds={"loc": "center left", "bbox_to_anchor": (1, 0.5)},
   )
```
